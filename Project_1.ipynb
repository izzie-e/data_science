```python
by_country.fillna(value=0, inplace=True)
date = by_country.date.str.split('-', expand=True)
date
```

```python
by_country['year'] = date[0]
by_country['month'] = date[1]
by_country['day'] = date[2]

by_country.year = pd.to_numeric(by_country.year)
by_country.month = pd.to_numeric(by_country.month)
by_country.day = pd.to_numeric(by_country.day)

by_country.date = pd.to_datetime(by_country.date)

by_country.head()
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x=by_country.date, y=by_country.daily_vaccinations)
plt.title('The Number of daily vaccinations dynamic')
plt.show()
```

```python
countries = by_country.groupby('country')['total_vaccinations'].max().sort_values(ascending= False)[:5].index

top_countries = pd.DataFrame(columns= by_country.columns)
for country in countries:
  top_countries = top_countries.append(by_country.loc[by_country['country'] == country])
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(top_countries['date'], 
             top_countries['daily_vaccinations_per_million'], 
             hue= top_countries['country'], ci= False)
plt.title('Vaccination procedure go on rapidly');
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(x=by_country.date, y=by_country.daily_vaccinations_per_million )
plt.title('The Number of daily_vaccinations_per_million ')
plt.show()
```
